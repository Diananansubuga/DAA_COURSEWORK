```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np

def swap(a, i, j):
  a[i], a[j] = a[j], a[i]

# Selection Sort
def selection_sort(arr):
  n = len(arr)
  for i in range(n):
    min_idx = i
    for j in range(i + 1, n):
      if arr[j] < arr[min_idx]:
        min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]
  return arr

# Insertion Sort
def insertion_sort(arr):
  for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and key < arr[j]:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j + 1] = key

# Merge Sort
def merge_sort(arr):
  if len(arr) <= 1:
    return arr
  mid = len(arr) // 2
  left = merge_sort(arr[:mid])
  right = merge_sort(arr[mid:])
  return merge(left, right)

def merge(left, right):
  result = []
  i = j = 0
  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      result.append(left[i])
      i += 1
    else:
      result.append(right[j])
      j += 1
  result.extend(left[i:])
  result.extend(right[j:])
  return result

# Generate random data for sorting
array_sizes = np.arange(100, 1000, 100)
execution_times_selection_sort = []
execution_times_insertion_sort = []
execution_times_merge_sort = []
for array_size in array_sizes:
  random_array = np.random.randint(1, 100, array_size)

  # Measure execution time for each sorting algorithm
  start_time = time.time()
  selection_sort(random_array.copy())
  selection_time = time.time() - start_time

  start_time = time.time()
  insertion_sort(random_array.copy())
  insertion_time = time.time() - start_time

  start_time = time.time()
  merge_sort(random_array.copy())
  merge_time = time.time() - start_time

  # Store execution times
  execution_times_selection_sort.append(selection_time)
  execution_times_insertion_sort.append(insertion_time)
  execution_times_merge_sort.append(merge_time)

# Plotting the execution time
plt.plot(array_sizes, execution_times_selection_sort, color='blue', label='Selection Sort')
plt.plot(array_sizes, execution_times_insertion_sort, color='green', label='Insertion Sort')
plt.plot(array_sizes, execution_times_merge_sort, color='orange', label='Merge Sort')
plt.xlabel('Array Size (n)')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time of Sorting Algorithms')
plt.legend()
plt.show()
```

The graph shows that all three sorting algorithms have execution times that increase with the array size, but at different rates. Selection sort has the slowest execution time, followed by insertion sort and merge sort. Merge sort is the fastest of the three algorithms.

The graph also shows that the rate at which the execution time increases with the array size is different for each algorithm. For selection sort, the execution time increases quadratically with the array size (O(n^2)). For insertion sort, the execution time increases quadratically in the worst case, but it is typically linear in practice (O(n)). For merge sort, the execution time increases logarithmically with the array size (O(n log n)).

The graph also shows that the difference in execution time between the different sorting algorithms becomes more pronounced as the array size increases.
